## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# find # of unique items
total_unique_items=len(purchase_data["Item ID"].unique())
print(total_unique_items)

# calculate Average price
Average_price=round(purchase_data["Price"].mean(),2)
print(Average_price)

# Calculate Number of Purchases
number_purchase=purchase_data["Purchase ID"].count()
print(number_purchase)

# Calculate total revenue
total_revenue=purchase_data["Price"].sum()
print(total_revenue)
```

* Run basic calculations to obtain number of unique items, average price, etc.

* Create a summary data frame to hold the results


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# find 'purchase count' by gender
purchase_male=gender_M["Purchase ID"].count()
purchase_female=gender_F["Purchase ID"].count()
purchase_others=gender_O["Purchase ID"].count()
#print(purchase_male)
#print(purchase_female)
#print(purchase_others)

# find 'Average purchase price' by gender
Avg_price_male=round(gender_M["Price"].mean(),2)
Avg_price_female=round(gender_F["Price"].mean(),2)
Avg_price_others=round(gender_O["Price"].mean(),2)
#print(Avg_price_male)
#print(Avg_price_female)
#print(Avg_price_others)

# Total purchase value by gender
Total_Value_M=round(gender_M["Price"].sum(),2)
Total_Value_F=round(gender_F["Price"].sum(),2)
Total_Value_O=round(gender_O["Price"].sum(),2)
#print(Total_Value_M)
#print(Total_Value_F)
#print(Total_Value_O)

# Calculate Normalize purchasing by gender
Norm_Total_M=round(Total_Value_M / purchase_male,2)
Norm_Total_F=round(Total_Value_F / purchase_female,2)
Norm_Total_O=round(Total_Value_O / purchase_others,2)
#print(Norm_Total_M)
#print(Norm_Total_F)
#print(Norm_Total_O)
```

```python
# Place data in summary table
purchase_gender=pd.DataFrame({"Purchase Count": [purchase_female, purchase_male, purchase_others],
                                    "Average Purchase Price": [Avg_price_female, Avg_price_male, 
                                                               Avg_price_others],
                              "Total Purchase Value":[Total_Value_F, Total_Value_M, Total_Value_O],
                              "Normalized Totals": [Norm_Total_F, Norm_Total_M, Norm_Total_O],
                              "Gender": ["Female", "Male", "Other / Non-Disclosed"]})
# set Gender as index
summary_purchase_gender=purchase_gender.set_index("Gender")
#summary_purchase_gender

# set format to the table
summary_purchase_gender["Average Purchase Price"]=summary_purchase_gender["Average Purchase Price"].map("${:,.2f}".format)
summary_purchase_gender["Total Purchase Value"]=summary_purchase_gender["Total Purchase Value"].map("${:,.2f}".format)
summary_purchase_gender["Normalized Totals"]=summary_purchase_gender["Normalized Totals"].map("${:,.2f}".format)
summary_purchase_gender
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# perchase count per player
purchase_pplayer=purchase_data.groupby("SN")
purchase_count_pplayer=purchase_pplayer["Purchase ID"].count()
#print(purchase_count_pplayer.head())

# Average purchase price per player
Avg_purchaseP_pplayer=round(purchase_pplayer["Price"].mean(),2)
#print(Avg_purchaseP_pplayer.head())

# Total purchase value per player
tot_purchaseV_pplayer=round(purchase_pplayer["Price"].sum(),2)
#print(tot_purchaseV_pplayer.head())

# Place all data in summary table
top_spender=pd.DataFrame({"Purchase Count": purchase_count_pplayer,
                         "Average Purchase Price": Avg_purchaseP_pplayer,
                         "Total Purchase Value": tot_purchaseV_pplayer})
#top_spender.head()
```

```python
# Sort table by total purchase value collumn in descending order
highest_purchase_value=top_spender.sort_values(["Total Purchase Value"], ascending=False)

# Add $ format
highest_purchase_value["Average Purchase Price"]=highest_purchase_value["Average Purchase Price"].map("${:,.2f}".format)
highest_purchase_value["Total Purchase Value"]=highest_purchase_value["Total Purchase Value"].map("${:,.2f}".format)

# show the table for the top 5 spenders
highest_purchase_value.head()
```

```python
#retrieve_data=purchase_data["Item ID", "Item Name", "Price"]
# look @ file for easy to manipulate data
# purchase_data.head()
```

```python
# group data by item name for calculation
item_name=purchase_data.groupby(["Item ID", "Item Name"])
item_name
```

```python
# Calculate purchase count per item
purchase_item_name_count=item_name["Item Name"].count()
#print(purchase_item_name_count)

# retrieve item price for each item name
# use either first() or last  to select only 1 data point since within Item group, prices are all the same 
item_name_price=item_name["Price"].first()
#print(item_name_price)

# Calculate Total purchase value for each item name
total_price_item_name=round(item_name["Price"].sum(),2)
#total_price_item_name

# Place all data in summary table (dataframe)
summary_item_name_purchase=pd.DataFrame({"Purchase Count": purchase_item_name_count,
                                        "Item Price": item_name_price,
                                        "Total Purchase Value": total_price_item_name})
#summary_item_name_purchase.head()
```

```python
# Sort the table with purchase count column in descending order
higest_purchase_count=summary_item_name_purchase.sort_values(["Purchase Count"],ascending=False)
higest_purchase_count.head()

# Format the table
higest_purchase_count["Total Purchase Value"]=higest_purchase_count["Total Purchase Value"].map("${:.2f}".format)
higest_purchase_count["Item Price"]=higest_purchase_count["Item Price"].map("${:.2f}".format)
higest_purchase_count.head()
```
